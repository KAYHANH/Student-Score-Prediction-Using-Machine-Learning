```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# Load dataset
data_path = 'C:/Users/Asus/Downloads/student-math-score-predictor/student_scores.csv'
data = pd.read_csv(data_path)

# Quick overview
print("Shape of the dataset:", data.shape)
display(data.head())
display(data.describe())
display(data.info())
```

```python
# -----------------------------------------------------------
#  Step 2: Exploratory Data Analysis (EDA)
# -----------------------------------------------------------

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Check for missing values
print("Missing values in each column:")
print(data.isnull().sum())

# 2. Plot distribution of all score columns
score_columns = [col for col in data.columns if "_score" in col]

plt.figure(figsize=(15, 10))
for idx, col in enumerate(score_columns):
    plt.subplot(3, 3, idx + 1)
    sns.histplot(data[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# 3. Correlation heatmap of scores
plt.figure(figsize=(10, 8))
sns.heatmap(data[score_columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Subject Scores")
plt.show()

# 4. Self-study hours vs score scatterplots
plt.figure(figsize=(15, 10))
for idx, col in enumerate(score_columns):
    plt.subplot(3, 3, idx + 1)
    sns.scatterplot(x='weekly_self_study_hours', y=col, data=data, hue='gender', palette='Set2', alpha=0.7)
    plt.title(f'Self-Study Hours vs {col}')
plt.tight_layout()
plt.show()

# 5. Boxplots to see effect of extracurricular activities
plt.figure(figsize=(15, 10))
for idx, col in enumerate(score_columns):
    plt.subplot(3, 3, idx + 1)
    sns.boxplot(x='extracurricular_activities', y=col, data=data, palette='Set3')
    plt.title(f'Activities vs {col}')
plt.tight_layout()
plt.show()
```

```python
# -----------------------------------------------------------
#  Step 3: Data Preprocessing
# -----------------------------------------------------------

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 1. Drop non-informative columns
data_cleaned = data.drop(columns=["id", "first_name", "last_name", "email"])

# 2. Encode categorical columns
label_encoders = {}
categorical_cols = ['gender', 'career_aspiration']

for col in categorical_cols:
    le = LabelEncoder()
    data_cleaned[col] = le.fit_transform(data_cleaned[col])
    label_encoders[col] = le

# 3. Define feature set (X) and target variables (Y)
score_columns = [col for col in data_cleaned.columns if '_score' in col]
X = data_cleaned.drop(columns=score_columns)
Y = data_cleaned[score_columns]  # Multi-output regression (predicting all scores)

# 4. Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 5. Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(" Preprocessing complete. X_train shape:", X_train_scaled.shape)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialize model
base_model = RandomForestRegressor(n_estimators=100, random_state=42)
model = MultiOutputRegressor(base_model)

# Train model
model.fit(X_train_scaled, Y_train)

# Predict
Y_pred = model.predict(X_test_scaled)

# Evaluation
def evaluate_model(true, pred):
    print("Evaluation Results:")
    for i, col in enumerate(score_columns):
        print(f"\n Subject: {col}")
        print(f"R Score: {r2_score(true.iloc[:, i], pred[:, i]):.4f}")
        print(f"MAE     : {mean_absolute_error(true.iloc[:, i], pred[:, i]):.2f}")
        print(f"RMSE    : {np.sqrt(mean_squared_error(true.iloc[:, i], pred[:, i])):.2f}")

evaluate_model(Y_test, Y_pred)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor

# Define and wrap models for multi-output regression
models = {
    "Linear Regression": MultiOutputRegressor(LinearRegression()),
    "Random Forest": MultiOutputRegressor(RandomForestRegressor(random_state=42)),
    "XGBoost": MultiOutputRegressor(XGBRegressor(random_state=42, verbosity=0))
}

trained_models = {}
for name, model in models.items():
    model.fit(X_train_scaled, Y_train)
    trained_models[name] = model
    print(f" {name} training complete")
```

```python
import joblib

# Save the best model (e.g., XGBoost)
joblib.dump(trained_models["XGBoost"], "xgboost_multioutput_model.pkl")
joblib.dump(scaler, "scaler.pkl")
print(" Model and scaler saved.")
```
